```python
import pandas as pd
```

```python
df = pd.read_csv('framingham.csv')
df
```

```python
# Data cleaning
df.isnull().sum()
```

```python
df = df.dropna(subset=['education'])
df = df.dropna(subset=['BPMeds'])
df = df.dropna(subset=['BMI'])
df = df.dropna(subset=['heartRate'])
df = df.dropna(subset=['glucose'])
df.isnull().sum()
```

```python
# Calculate the mean of 'cigsPerDay'
mean_cigsPerDay = df['cigsPerDay'].mean()
mean_chol = df['totChol'].mean()
# Replace null values in 'cigsPerDay' with the mean
df[['cigsPerDay']] = df[['cigsPerDay']].fillna(mean_cigsPerDay)
df[['totChol']] = df[['totChol']].fillna(mean_chol)
mean_cigsPerDay
```

```python
from sklearn.model_selection import train_test_split
```

# Separate the feature and target 

```python
X = df.drop('TenYearCHD', axis = 1)
y = df['TenYearCHD']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=413) # The random state is set to 413 to ensure reproducibility
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=3)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred
```

```python
accuracy = (y_pred == y_test).mean()
accuracy
```

```python
from sklearn import tree 
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
tree.plot_tree(dt, filled=True)
plt.show()
```
