```python
import pandas as pd
melbourne_data = pd.read_csv('melb_data.csv')
melbourne_data.head()
```

```python
melbourne_data.describe()
```

```python
melbourne_data['Price'] = melbourne_data['Price']*.65
# This changes the price so it becomes to USD
```

```python
import seaborn as sns
sns.histplot(melbourne_data['Price'], kde=True)
```

```python
melbourne_data.YearBuilt.fillna(melbourne_data.YearBuilt.mean(), inplace=True)
melbourne_data.BuildingArea.fillna(melbourne_data.BuildingArea.mean(), inplace=True)
melbourne_data.Car.fillna(melbourne_data.Car.mean(), inplace=True)
```

```python
y = melbourne_data.Price
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude', 'Postcode', 'BuildingArea', 'YearBuilt', "Car", "Distance"]
X = melbourne_data[features]
```

```python
melbourne_data.isnull().sum()
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=1)
dt.fit(X,y)
```

```python
print(dt.predict(X.head()))
y.head()
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = dt.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

In the previous model we trained the data on the data we also tried to predict. 

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.2)

```

```python
dt2 = DecisionTreeRegressor(random_state=1, max_depth=5) 
dt2.fit(X_train, y_train)
```

```python
predicted_home_prices2 = dt2.predict(x_test)
mean_absolute_error(y_test, predicted_home_prices2)
```

```python
# We are now gonna compar ethe 
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, X_train, x_test, y_train, y_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(X_train, y_train)
    preds_val = model.predict(x_test)
    mae = mean_absolute_error(y_test, preds_val)
    return(mae)
```

```python
max_my_mae = 0
for max_leaf_nodes in range(400, 500):
    my_mae = get_mae(max_leaf_nodes, X_train, x_test, y_train, y_test)
    # print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
    if max_my_mae == 0 or my_mae < max_my_mae:
        max_my_mae = my_mae
        best_max_leaf_nodes = max_leaf_nodes
print("Best max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(best_max_leaf_nodes, max_my_mae))
```
